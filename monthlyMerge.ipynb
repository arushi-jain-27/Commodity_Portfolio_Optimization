```python
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('predictors.xlsx')
```

```python
df['date'] = df['date'].dt.to_period('M')
```

```python
df.head()
```

```python
df_new = df.groupby(['date'], as_index=False).mean() # Take mean for each month
```

```python
df_new = df_new.set_index('date')
```

```python
df_new = df_new.interpolate(method ='linear', limit_direction ='forward') # Interpolate NaN
```

```python
df_new.to_excel('predictors_monthly.xlsx')
```
